```python
# Diferentes Ángulos
def Angulo(fila):
    if fila.loc_y < 0:
        zona = "Sin angulo"
    elif abs(fila.loc_x) <= (fila.loc_y*0.26796):
        zona = "0-15"
    elif abs(fila.loc_x) <= (fila.loc_y*0.57732):
        zona = "15-30"
    elif abs(fila.loc_x) <= (fila.loc_y*1):
        zona = "30-45"
    elif abs(fila.loc_x) <= (fila.loc_y*1.73204):
        zona = "45-60"
    elif abs(fila.loc_x) <= (fila.loc_y*3.73204):
        zona = "60-75"
    else:
        zona = "75-90"
    return zona
```

```python
# Agregamos la variable angulo
tabla['angulo']=tabla.apply(Angulo, axis=1)
tabla.angulo
```
